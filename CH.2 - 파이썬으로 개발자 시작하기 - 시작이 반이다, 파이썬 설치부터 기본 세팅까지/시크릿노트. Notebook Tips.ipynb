## 1. 노트북의 기준이 되는 셀(cell)

### 셀(cell)이란?
notebook에서 cell은 기준이 되는 하나의 블록(block)입니다. 
그래서 저희가 코드를 치는 곳도 셀이라고 부릅니다.

### 셀의 종류
notebook의 위에 보면, 아래 이미지와 같은 `code`라고 적힌 부분이 있습니다.
이 부분을 클릭하시면, `code`/`markdown`등과 같은 선택 도구가 뜨는데요. 여기에 뜨는 게 `셀의 타입`입니다. 



대표적인 셀의 타입은 `code`이며, 저희가 파이썬 코딩을 할 수 있는 셀을 의미합니다. 두번째로 대표적인 타입은 `markdown`이며, 마크다운은 실제로 개발자들이 문서를 적을 때 많이 사용하는 문법 도구입니다. 

이 두개는 간단하게 `code`는 **실제 코딩을 하는 곳**, `markdown`은 **텍스트를 적는곳** 이라고 생각하시면 좋을 것 같습니다.



```python
# 코드를 기입하는 셀입니다.
```

> 마크다운을 기입할 수 있는 셀입니다.

## 2. 노트북을 유용하게 사용 할 수 있는 단축키

notebook은 간단한 단축키들을 갖고 있습니다. 셀 단축키는 code나 markdown에서 편집중인 상태에선 사용할 수 없고, 셀을 선택하거나, 실행한 후에 사용할 수 있습니다.

**단축키를 사용할 수 있는 상태(Command Mode)**


**단축키를 사용할 수 없는 상태(Edit Mode)**


Edit Mode상태에서 <kbd>Esc</kbd>를 누르면, 단축키를 사용할 수 있는 상태로 들어갈 수 있습니다.

### 단축키 보기
아무런 셀이 선택되지 않은 상태에서 <kbd> h </kbd>를 누르면, 모든 단축키를 볼 수 있습니다.

### 셀 관련 유용한 단축키
<kbd> Ctrl </kbd> + <kbd> Enter </kbd> : 현재 선택된 셀의 코드를 실행합니다. 

<kbd> Alt </kbd> + <kbd> Enter </kbd> : 현재 선택된 셀의 코드를 실행합니다. 그리고, 아래에 새로운 셀을 만듭니다.

<kbd> A </kbd> : 현재 셀 위에 새로운 셀을 만듭니다.

<kbd> B </kbd> : 현재 셀 아래에 새로운 셀을 만듭니다.

<kbd> D </kbd> x 2 : <kbd> D </kbd>를 두번 탭해서 현재 셀을 삭제합니다.
